```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
MMM = pd.read_csv("Data/Data Given/MMM.csv").dropna()
MMMDataFrom2014to2018 = MMM[50*5*5:]
MMMDataFrom2014to2018 = MMMDataFrom2014to2018.drop(columns="Volume")

HON = pd.read_csv("Data/Data Given/HON.csv").dropna()
HONDataFrom2014to2018 = HON[50*5*5:]
HONDataFrom2014to2018 = HONDataFrom2014to2018.drop(columns="Volume")

SYF = pd.read_csv("Data/Data Given/SYF.csv").dropna()
SYFDataFrom2014to2018 = SYF
SYFDataFrom2014to2018 = SYFDataFrom2014to2018.drop(columns="Volume")

BAYZF = pd.read_csv("Data/Data Given/BAYZF.csv").dropna()
BAYZFDataFrom2014to2018 = BAYZF[50*5*5:]
BAYZFDataFrom2014to2018 = BAYZFDataFrom2014to2018.drop(columns="Volume")
```

```python
MMMDates = MMMDataFrom2014to2018["Date"]
MMMClosingValues = MMMDataFrom2014to2018["Close"]

MMMClosingValues[1250+0]
MMMXAxis = np.arange(400)
MMMYAxis = np.zeros(400)
temp = 0
counter = 0
for i in range(len(MMMDates)):
    if counter == 400:
        break
    if np.abs(MMMClosingValues[1250+i] - temp) > np.abs(0.01*MMMClosingValues[1250+i]):
        temp = MMMClosingValues[1250+i]
        MMMYAxis[counter] = temp
        counter += 1

plt.figure(1)
plt.plot(MMMXAxis,MMMYAxis)
plt.ylabel("Cost per Share (USD)")
plt.xlabel("Units of Time")
plt.title("Changes in 3M Stock")
```

```python
HONDates = HONDataFrom2014to2018["Date"]
HONClosingValues = HONDataFrom2014to2018["Close"]

HONClosingValues[1250+0]
HONXAxis = np.arange(375)
HONYAxis = np.zeros(375)
temp = 0
counter = 0
for i in range(len(HONDates)):
    if counter == 375:
        break
    if np.abs(HONClosingValues[1250+i] - temp) > np.abs(0.01*HONClosingValues[1250+i]):
        temp = HONClosingValues[1250+i]
        HONYAxis[counter] = temp
        counter += 1

plt.figure(2)
plt.plot(HONXAxis,HONYAxis)
plt.ylabel("Cost per Share (USD)")
plt.xlabel("Units of Time")
plt.title("Changes in Honeywell Stock")
```

```python
SYFDates = SYFDataFrom2014to2018["Date"]
SYFClosingValues = SYFDataFrom2014to2018["Close"]

SYFXAxis = np.arange(525)
SYFYAxis = np.zeros(525)
temp = 0
counter = 0
for i in range(len(SYFDates)):
    if counter == 525:
        break
    if np.abs(SYFClosingValues[i] - temp) > np.abs(0.01*SYFClosingValues[i]):
        temp = SYFClosingValues[i]
        SYFYAxis[counter] = temp
        counter += 1

plt.figure(3)
plt.plot(SYFXAxis,SYFYAxis)
plt.ylabel("Cost per Share (USD)")
plt.xlabel("Units of Time")
plt.title("Changes in Synchrony Stock")
```

```python
BAYZFDates = BAYZFDataFrom2014to2018["Date"]
BAYZFClosingValues = BAYZFDataFrom2014to2018["Close"]

BAYZFClosingValues[1250+0]
BAYZFXAxis = np.arange(450)
BAYZFYAxis = np.zeros(450)
temp = 0
counter = 0
for i in range(len(BAYZFDates)):
    if counter == 450:
        break
    if np.abs(BAYZFClosingValues[1250+i] - temp) > np.abs(0.01*BAYZFClosingValues[1250+i]):
        temp = BAYZFClosingValues[1250+i]
        BAYZFYAxis[counter] = temp
        counter += 1

plt.figure(4)
plt.plot(BAYZFXAxis,BAYZFYAxis)
plt.ylabel("Cost per Share (USD)")
plt.xlabel("Units of Time")
plt.title("Changes in Bayer Stock")
```

```python
((pd.DataFrame([BAYZFXAxis,BAYZFYAxis])).transpose()).to_csv("SlimmedBAYZF.csv",header=None, index=None)
((pd.DataFrame([HONXAxis,HONYAxis])).transpose()).to_csv("SlimmedHON.csv",header=None, index=None)
((pd.DataFrame([SYFXAxis,SYFYAxis])).transpose()).to_csv("SlimmedSYF.csv",header=None, index=None)
((pd.DataFrame([MMMXAxis,MMMYAxis])).transpose()).to_csv("SlimmedMMM.csv",header=None, index=None)
```
